```python
#creates a categorical region list out of countries

#Numbers for regions

#1: Central Europe
#2: Mediterranean
#3: Northern Europe
#4: Eastern Europe
#5: MENA (Middle East and North Africa)
#0: Not in the analysis

international_region = []

for i in range(len(all_recipes.country)):
    if all_recipes.country[i] == 'France' or all_recipes.country[i] == 'Germany' or all_recipes.country[i] == 'Netherlands' or all_recipes.country[i] == 'Austria' or all_recipes.country[i] == 'Switzerland' or all_recipes.country[i] == 'Belgium':
        international_region.append(1)
    elif all_recipes.country[i] == 'Italy' or all_recipes.country[i] == 'Spain' or all_recipes.country[i] == 'Portugal' or all_recipes.country[i] == 'Turkey':
        international_region.append(2)
    elif all_recipes.country[i] == 'UK-and-Ireland' or all_recipes.country[i] == 'Scandinavia':
        international_region.append(3)
    elif all_recipes.country[i] == 'Eastern-Europe':
        international_region.append(4)
    elif all_recipes.country[i] == 'North-African' or all_recipes.country[i] == 'Lebanon' or all_recipes.country[i] == 'Iran' or all_recipes.country[i] == 'Israel':
        international_region.append(5)
    else:
        international_region.append(0)
```

```python
#attaches the international region list to the data frame

all_recipes['international_region'] = international_region

recipes_mena_europe = all_recipes[all_recipes.international_region != 0]

recipes_mena_europe.index = range(len(recipes_mena_europe))
```

```python
#Creates a list of lists for the recipes

elements = []
alist=[]

for i in range(len(recipes_mena_europe)):
    elements.append([recipes_mena_europe.iloc[i,1],recipes_mena_europe.iloc[i,2],
                         recipes_mena_europe.iloc[i,3],recipes_mena_europe.iloc[i,4],
                         recipes_mena_europe.iloc[i,5],recipes_mena_europe.iloc[i,6],
                         recipes_mena_europe.iloc[i,7],recipes_mena_europe.iloc[i,8],
                         recipes_mena_europe.iloc[i,9],recipes_mena_europe.iloc[i,10],
                         recipes_mena_europe.iloc[i,11],recipes_mena_europe.iloc[i,12],
                         recipes_mena_europe.iloc[i,13],recipes_mena_europe.iloc[i,14],
                         recipes_mena_europe.iloc[i,15],recipes_mena_europe.iloc[i,16],
                         recipes_mena_europe.iloc[i,17],recipes_mena_europe.iloc[i,18],
                         recipes_mena_europe.iloc[i,19],recipes_mena_europe.iloc[i,20],
                         recipes_mena_europe.iloc[i,21],recipes_mena_europe.iloc[i,22],
                         recipes_mena_europe.iloc[i,23],recipes_mena_europe.iloc[i,24],
                         recipes_mena_europe.iloc[i,25],recipes_mena_europe.iloc[i,26],
                         recipes_mena_europe.iloc[i,27],recipes_mena_europe.iloc[i,28],
                         recipes_mena_europe.iloc[i,29],recipes_mena_europe.iloc[i,30]])
    alist.append(elements[i])
```

```python
#defines a matrix to insert similarity values

similarity_matrix = [[0 for x in range(975)] for x in range(975)] 

#The double for loop below calculates the Jaccard similarity between the North European and Central European recipes.

for i in range(2010,2985):
    j = i + 1
    set1 = Set(alist[i])
    if j > 2985:
        break
    for j in range(2010,2985):
        set2 = Set(alist[j])
        union = len(set1 | set2)
        intersection = len(set1 & set2)
        similar = (float(intersection) / float(union))
        similarity_matrix[i-2010][j-2010] = similar
```

```python
#Dendrogram of the Jaccard similarity matrix

Z = hr.linkage(similarity_matrix, method='complete', metric='jaccard')

fig = plt.figure(figsize=(10,10))
T = hr.dendrogram(Z,color_threshold=1, leaf_font_size=4)
fig.show()
```

```python
cg = sns.clustermap(similarity_matrix, method="complete", figsize=(13,13), xticklabels=False)
print cg.dendrogram_col.reordered_ind
```
