# Analyzing hotel ratings on Tripadvisor

In this homework we will focus on practicing two techniques: web scraping and regression. For the first part, we will build upon the sample code from the Lecture and attempt to get some basic information for each hotel. Then, we will fit a regression model on this information and try to analyze it.   

One of the main disadvantages of scraping a website instead of using an API is that, without any notice, the website may change its layout and render our code useless. Something like that happened in our case. Tripadvisor changed the layout of the buttons that we use to navigate between the different pages of the results. This was the main reason people were having problem with executing the code.

** Task 3 (20 pts) **

Now, we will use regression to analyze this information. First, we will fit a linear regression model that predicts the average rating. For example, for the hotel above, the average rating is

$$ \text{AVG_SCORE} = \frac{1*31 + 2*33 + 3*98 + 4*504 + 5*1861}{2527}$$

Use the model to analyze the important factors that decide the $\text{AVG_SCORE}$.

** Task 4 (30 pts) **

Finally, we will use logistic regression to decide if a hotel is _excellent_ or not. We classify a hotel as _excellent_ if more than **60%** of its ratings are 5 stars. This is a binary attribute on which we can fit a logistic regression model. As before, use the model to analyze the data.

````python
#This short code converts BeautifulSoup web page links into strings so that I can open them later.

string_html = []

for i in range(len(hrefs)):
    string_html.append(unicode.join(u'\n',map(unicode,hrefs[i])))

print len(string_html)
````

````python
#This short code parses out the hotel links from string html created above.

clean_link = []

for i in range(len(string_html)):
    s = string_html[i]
    clean_link.append("http://www.tripadvisor.com/" + s.partition('<a target="_blank" href="/')[-1].partition('"')[0])
````

````python
#This short code retrieves information for all hotels (77 in total for Boston).

hotel_info = []

for i in range(len(clean_link)):
    hotel_info.append(urlopen(clean_link[i]).read())
````

````python
#Here I retrieve three pieces of ratings as requsted in the assignment. The code finds all the instances that we need. In the last line
#I used [1:7] to get the instances of summary ratings that we need.

traveler_rating = []
review_type = []
sum_rating = []
temp = []

for i in range(len(hotel_info)):
    traveler_rating.append(re.findall(r'\compositeCount">(.*)\</span>', hotel_info[i]))
    review_type.append(re.findall(r'\class="value">(.*)\</div>', hotel_info[i]))
    temp = re.findall(r'\" src="http://e2.tacdn.com/img2/x.gif" alt="(.*)\ of 5 stars">', hotel_info[i])
    sum_rating.append(temp[1:7])
````

````python
#Here I convert the ratings that are strings into float numbers. I wasn't able to do the same process for the summary ratings, so I needed
#them out of my analysis. You can see my code below. Hopefully, I get some partial credit :) Python gives the following error:
#invalid literal for int() with base 10: '4.5' I think it's because of the . in between numbers, I solved this problem for the first two
#ratings by removing the comma "," as you can see with the replace command.

traveler_float_rating = []
review_float_type = []
sum_float_rating = []

for i in range(len(traveler_rating)):
    for j in range(5):
        if "," in traveler_rating[i][j]:
            traveler_rating[i][j] = traveler_rating[i][j].replace(",","")
    traveler_float_rating.append(map(float, traveler_rating[i]))

for i in range(len(traveler_rating)):
    for j in range(4):
        if "," in review_type[i][j]:
            review_type[i][j] = review_type[i][j].replace(",","")
    review_float_type.append(map(float, review_type[i]))
    
'''

for i in range(len(sum_rating)):
    for j in range(len(sum_rating[i])):
       if "," in sum_rating[i][j]:
            sum_rating[i][j] = sum_rating[i][j].replace(",","")
    sum_int_rating.append(map(int, sum_rating[i]))
    

'''
````

````python
#Here I calculate the average score for the linear regression analysis, as requested in the assignment.

average_score = []
score = 0.0
number_of_reviewers = 0.0

for i in range(len(traveler_float_rating)):
    score = 0.0
    number_of_reviewers = 0.0
    a = 5
    for j in range(5):
        score += a*traveler_float_rating[i][j] 
        number_of_reviewers += traveler_float_rating[i][j]
        a -= 1
    average_score.append(score/number_of_reviewers)
````

````python
#We can analyze how traveler type influences the rating.
#Hypothesis: Travelers that belong to different categories rates the hotel differently, since their perception depends on their traveler status.

families = []
couples = []
solo = []
business = []

for i in range(len(review_float_type)):
    families.append(review_float_type[i][0])
    couples.append(review_float_type[i][1])
    solo.append(review_float_type[i][2])
    business.append(review_float_type[i][3])
````

````python
#Below you can see the simple linear regression results. Results are statistically significant. Solo travellers tend to give higher ratings.

print "RESULTS FOR FAMILIES"

model = sm.OLS(average_score, families)
results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params

print "RESULTS FOR COUPLES"

model = sm.OLS(average_score, couples)
results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params

print "RESULTS FOR SOLO TRAVELERS"

model = sm.OLS(average_score, solo)
results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params

print "RESULTS FOR BUSINESS TRAVELERS"

model = sm.OLS(average_score, business)
results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params
````

````python
#Here I categorized the hotels as excellent or not by assigning them the values "1" for excellent and "0" for not excellent.

excellent_hotels = []

for i in range(len(traveler_float_rating)):
    if traveler_float_rating[i][1] / number_of_reviewers  >= 0.6:
        excellent_hotels.append(1)
    else:
        excellent_hotels.append(0)
````

````python
#Below you can see my logistic regression results. Here you can see that it is more likely for solo travellers to categorize hotels as
#excellent.

print "RESULTS FOR FAMILIES"

logit = sm.Logit(excellent_hotels, families)
result = logit.fit()
print result.summary()

print "RESULTS FOR COUPLES"

logit = sm.Logit(excellent_hotels, couples)
result = logit.fit()
print result.summary()

print "RESULTS FOR SOLO TRAVELERS"

logit = sm.Logit(excellent_hotels, solo)
result = logit.fit()
print result.summary()

print "RESULTS FOR BUSINESS TRAVELERS"

logit = sm.Logit(excellent_hotels, business)
result = logit.fit()
print result.summary()
````
