## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
from sklearn import preprocessing
import numpy as np
import sklearn.metrics as metrics
from collections import Counter 
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as sk_data
import sklearn.metrics as metrics
from sklearn.cluster import KMeans

LATITUDE = []
LONGITUDE = []

for i in range(len(boston.LATITUDE)):
    LATITUDE.append(boston.LATITUDE[i])

for i in range(len(boston.LONGITUDE)):
    LONGITUDE.append(boston.LONGITUDE[i])
    
LONGITUDE_scaled = preprocessing.scale(LONGITUDE)
LATITUDE_scaled = preprocessing.scale(LATITUDE)
    
geodata = zip(LONGITUDE,LATITUDE)

geo_scaled = preprocessing.scale(geodata)

def evaluate_clusters(geo_scaled,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(geo_scaled)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
    
#neighborhood_services_district = Counter(boston.neighborhood_services_district)

#print neighborhood_services_district

    
#Five clusters can be chosen, since a lot less variance is explained after five clusters, there is no break in the line after 5.
```

```python
evaluate_clusters(geo_scaled,10)
```

```python
evaluate_clusters(geo_scaled,20)
```

```python
kmeans = KMeans(init='k-means++', n_clusters=5, n_init=10)
kmeans.fit_predict(geo_scaled)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

print "The total error of the clustering is: ", error
print '\nCluster labels'
print labels
print '\n Cluster Centroids'
print centroids


fig, ax = plt.subplots(1,1,figsize=(10,10))
cmap = sns.color_palette("bright", n_colors=5)
cmap.extend(sns.hls_palette(n_colors=6, h=0.7, l=0.4, s=0.8))

for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    plt.scatter(LONGITUDE_scaled[point_indices], LATITUDE_scaled[point_indices], s=25, alpha=0.2, c=cmap[i], marker='+')

```

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!)
```python
#We can check whether people from certain districts make calls at certain times of the day using a clustering mechanism. 
#We can also check whether complaints from certain regions (latitudes/longitudes) are opened or closed at a certain time. 
#We can use the OPEN_DT, CLOSED_DT, and LOCATION_ZIPCODE variables in the dataframe for this purpose.
#We can then find whether there are similar regions in this respect.

import datetime
import time

time_open = []
time_split_open = []
time_decimal_open = []
time_closed = []
time_split_closed = []
time_decimal_closed = []
time_spent = []
zipcode = []
location = []

#create a new data frame that does not include the NaN values for CLOSED_DT and LOCATION_ZIPCODE variables

boston_closed = boston[boston.CLOSED_DT.notnull()]
boston_closed_zipcode = boston_closed[boston_closed.LOCATION_ZIPCODE.notnull()]



boston_closed_zipcode.index = range(len(boston_closed_zipcode))

#append OPEN_DT information to a list and convert time to seconds

for i in range(len(boston_closed_zipcode.OPEN_DT)):
        time_open.append(boston_closed_zipcode.OPEN_DT[i])
    
for i in range (len(time_open)):
    time_split_open.append(time_open[i].split(" "))
    
time_hours_open = [i[1] for i in time_split_open]

for i in range(len(time_hours_open)):
    (h, m, s) = time_hours_open[i].split(":")
    time_decimal_open.append(float(h) * 3600 + float(m) * 60 + float(s))


#append CLOSED_DT information to a list and convert time to seconds

for i in range(len(boston_closed_zipcode.CLOSED_DT)):
    time_closed.append(boston_closed_zipcode.CLOSED_DT[i])
    
for i in range (len(time_closed)):
    time_split_closed.append(time_closed[i].split(" "))
    
time_hours_closed = [i[1] for i in time_split_closed]

for i in range(len(time_hours_closed)):
    (h, m, s) = time_hours_closed[i].split(":")
    time_decimal_closed.append(float(h) * 3600 + float(m) * 60 + float(s))

#converting latitude and longitude into a single variable according to geospatial indexing

for i in range(len(boston_closed_zipcode.LATITUDE)):
    location.append((boston_closed_zipcode.LATITUDE[i]+90)*180+boston_closed_zipcode.LONGITUDE[i])

#append LOCATION_ZIPCODE information to a list

for i in range(len(boston_closed_zipcode.LOCATION_ZIPCODE)):
    zipcode.append(float(boston_closed_zipcode.LOCATION_ZIPCODE[i]))

#scaling the variables

time_decimal_open_scaled = preprocessing.scale(time_decimal_open)

time_decimal_closed_scaled = preprocessing.scale(time_decimal_closed)

location_scaled = preprocessing.scale(location)

new_cluster_data = zip(time_decimal_open_scaled, location_scaled)

new_cluster_data2 = zip(time_decimal_closed_scaled, location_scaled)

new_cluster_data3 = zip(time_decimal_open_scaled, zipcode)

#Four clusters can be chosen, since a lot less variance is explained after four clusters, there is no break in the line after 4.
```
