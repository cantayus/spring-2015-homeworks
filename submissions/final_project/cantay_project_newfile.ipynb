```python
#Opens the policy page of each country and stores it in a list.
country_page = []
for i in range(len(all_countries)):
    country_page.append(BeautifulSoup(urlopen("http://www.iea.org/policiesandmeasures/renewableenergy/?country=%s" % all_countries[i]).read()))
    print i,
```

```python
#Creates a list of lists. In each list there are the policy links from a country's web page. Each country is a list in a list of lists.
policy_links = []
string_html = []
clean_link = []
complete_link = []

for i in range(len(country_page)):
    policy_links.append(country_page[i].findAll('a', href = True))
    string_html.append(unicode.join(u'\n',map(unicode,policy_links[i])))        
    clean_link.append(re.findall(r'\<a href="/policiesandmeasures(.*)\">', string_html[i]))
    sub_link_list = []
    for j in range(1,len(clean_link[i])-1):
        temp = "http://www.iea.org/policiesandmeasures/" + clean_link[i][j]
        sub_link_list.append(temp)
    complete_link.append(sub_link_list[0:(len(clean_link[i])-1)])
```

```python
#Opens every policy web page for every country and saves it under the policy_info list. Takes some time.
policy_info = []

for i in range(len(complete_link)):
    sub_policy_list = []
    for j in range(len(complete_link[i])):
        txt = urlopen(complete_link[i][j]).read()
        sub_policy_list.append(txt)
    policy_info.append(sub_policy_list[0:(len(complete_link[i]))])
    print i,
```

```python
#Stores the beginning date and type of policies in two separate lists. Policies that impact multiple renewable energy sources and that
#are national are taken into consideration. A separate analysis can be made for state-level policies for specific types of energy.\n",
      
policy_type = []
policy_begin = []

for i in range(len(policy_info)):
    sub_policy_type_list = []
    sub_policy_begin_list = []
    for j in range(len(policy_info[i])):
        if '<tr><td>Jurisdiction:</td><td>National</td></tr>' and 'Multiple RE Sources' in policy_info[i][j]:
            poltype = re.findall(r'\<tr><td>Policy Type:</td><td>(.*)\</td></tr>', policy_info[i][j])
            sub_policy_type_list.append(poltype)
            polbegin = re.findall(r'\<tr><td>Date Effective:</td><td>(.*)\</td></tr>', policy_info[i][j])
            sub_policy_begin_list.append(polbegin)
    policy_type.append(sub_policy_type_list[0:(len(policy_info[i]))])
    policy_begin.append(sub_policy_begin_list[0:(len(policy_info[i]))])

counter = 0
for i in range(len(policy_type)):
    for j in range(len(policy_type[i])):
        counter += 1
print counter
counter = 0
for i in range(len(policy_begin)):
    for j in range(len(policy_begin[i])):
        counter += 1
print counter

print len(policy_begin)
print len(policy_type)
```

```python
#Stores the end date of policies. Since there is some discrepancy between the policy pages, I needed to create if conditions to get the\n",
#end dates.

policy_end = []

for i in range(len(policy_info)):
    sub_policy_end_list = []
    for j in range(len(policy_info[i])):
        if '<tr><td>Jurisdiction:</td><td>National</td></tr>' and 'Multiple RE Sources' in policy_info[i][j]:
            if '<tr><td>Policy status:</td><td>Ended</td></tr>' in policy_info[i][j]:
                if '<tr><td>Date Ended:</td><td>' in policy_info[i][j]:
                    polend = re.findall(r'\<tr><td>Date Ended:</td><td>(.*)\</td></tr>', policy_info[i][j])
                    sub_policy_end_list.append(polend)
                else:
                    polend = re.findall(r'\<tr><td>Date Effective:</td><td>(.*)\</td></tr>', policy_info[i][j])
                    sub_policy_end_list.append(polend)
            else:
                polend = ['2014']
                sub_policy_end_list.append(polend)
    policy_end.append(sub_policy_end_list[0:(len(policy_info[i]))])

counter = 0
for i in range(len(policy_end)):
    for j in range(len(policy_end[i])):
        counter += 1
print counter

print len(policy_end)

print len(policy_end[0])
print len(policy_begin[0])
print len(policy_type[0])

print policy_end
```

```python
#Calculates the Jaccard similarity between countries. I took the averages of the binary variables for the 1990-2014 period and then compared
#109 sets of averages with other sets using the Jaccard similarity procedure.
data_jaccard = []
i = 0
while (i < len(data)):
    tmp = None
    tmp = data [["Funds to Sub-National Governments", "Infrastructure Investments", "Procurement Rules", "RD&D Funding",
                 "Feed-in Tariffs or Premiums", "Grants and Subsidies", "Loans", "Tax Relief", "Taxes", "User Charges",
                 "Green Certificates", "White Certificates", "Information and Education", "Institutional Creation", "Strategic Planning",
                 "Auditing", "Codes and Standards", "Monitoring", "Obligation Schemes", "Other Mandatory Requirements",
                 "Demonstration Project", "Research Programme"]][i:i+25].mean()
    data_jaccard.append(tmp)
    i += 25
```

```python
#defines a matrix to insert similarity values

similarity_matrix = [[0 for x in range(109)] for x in range(109)]
      
#The double for loop below calculates the Jaccard similarity for all countries.
      
for i in range(len(data_jaccard)):
    j = i + 1
    set1 = Set(data_jaccard[i])
    if j > 109:
        break
    for j in range(len(data_jaccard)):
        set2 = Set(data_jaccard[j])
        union = len(set1 | set2)
        intersection = len(set1 & set2)
        similar = (float(intersection) / float(union))
        similarity_matrix[i][j] = similar
```

```python
#Dendrogram of the Jaccard similarity matrix
      
Z = hr.linkage(similarity_matrix, method='complete', metric='jaccard')
      
fig = plt.figure(figsize=(30,30))
T = hr.dendrogram(Z,color_threshold=1, leaf_font_size=15, labels = all_countries)
fig.show()
```
