Now, let's plot (preferably with horizontal bars) the counts for all the `CASE_TITLE`s that occur more than 250 times in the dataset. **(5 pts)**

````python
case_title_values_list = case_title_values.most_common()

indices = [i[1] for i in case_title_values_list] #splits the list of lists into lists with column names and number of occurrences
columns = [i[0] for i in case_title_values_list]

indices_needed = []
columns_needed = []

for i in range(len(indices)): #for loop to create an array that contains number of occurrences > 250
    if indices[i] > 250:
        indices_needed.append(indices[i])
    else:
        break

data_needed = case_title_values_list[0:i]
        
for j in range(i): #for loop to create an array that contains the names of occurrences
    columns_needed.append(columns[j])
    
def plot9():
    data = data_needed
    N = len( data )
    x = np.arange(1, N+1)
    y = [ num for (s, num) in data ]
    labels = columns_needed
    width = 0.5
    bar1 = plt.bar(x, y, width, color="y" )
    plt.ylabel( 'Occurrence' )
    plt.xticks(x + width/2.0, labels, rotation='vertical')
    plt.show()

plot9() #draws the bar graph
````
